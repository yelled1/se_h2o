```python
import h2o

h2o.init()
```

```python
url = "https://raw.githubusercontent.com/DarrenCook/h2o/bk/datasets/"
iris = h2o.import_file(url+"iris_wheader.csv")
```

```python
train, test = iris.split_frame([.8])
y_col = 'class'
x_cols = iris.col_names
x_cols.remove(y_col)
```

```python
mod_rforest = h2o.estimators.random_forest.H2ORandomForestEstimator()
mod_rforest.train(x_cols, y_col, train, model_id='randforest_iris')
predRF = mod_rforest.predict(test)
```

```python
mod_rforest
```

```python
predRF.summary()
```

```python
mod_rforest.confusion_matrix(test)
```

```python
mod_gbm = h2o.estimators.H2OGradientBoostingEstimator()
mod_gbm.train(x_cols, y_col, train, model_id='mod_gbm')
pred_gbm = mod_gbm.predict(test)
```

```python
mod_gbm.summary()
```

```python
mod_gbm.confusion_matrix(test)
```
